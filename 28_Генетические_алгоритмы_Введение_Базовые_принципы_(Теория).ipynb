##  Естественный отбор в природе

Vvedenie-v-evolyutsiyu-i-estestvennyj-otbor.jpg

Как и в основе построения нейронной сети заложена модель работы живого нейрона, так и в основе идеи генетического алгоритма лежит теория эволюции живых систем.

Естественный отбор в природе состоит в том, что наименее приспособленные организмы погибают раньше и в больших количествах, не дав нужного потомства для продолжения своего вида, а наиболее приспособленные - выживают и дают потомство. Потомство выживших в среднем уже лучше приспособленно к окружающему миру, но опять же, среди них есть кто лучше приспособлен, а кто хуже, и далее отбор работает уже среди потомков.







image.png

Если вчитаться в принцип естественного отбора, то можно увидить там несколько понятий:
- особь;
- популяция;
- отбор;
- мутация;
- потомство;

Если мы смотрим, то в целом все понятно:

`Особь` - это представитель когого-то вида организмов.

`Популяция` - это сообщество данных особей, которые способны к скрещиванию друг с другом, что важно, так как для естественного отбора нужно потомство.


`Отбор` - это условия внешней среды (температура, влажность, освещенность, концентрация кислорода ...), которые определяют глобальные парамметры для жизнидеятельности любого организма способного адаптироваться и выжить в данной среде. 

`Мутация` - как известно, дети унаследуют большую, но все-таки часть генетической информации от родителей, а часть генов возникает случайным способом. Как гласит википедия - "... механизмы восстановления ДНК могут вносить мутационные ошибки при восстановлении физического повреждения молекулы. Восстановление, даже с мутацией, является более важным для выживания, чем восстановление точной копии ...". То есть в мутировших генах возникает новая информация, которая может помочь (или нет) в изменчивасти для естественного отбора  среди потомства.

`Потомство` - это дети выживших особей ранней популяции, которые получили часть ген от выживших особей, а часть от мутации. Выжившие особи с потомством образуют новую популяцию.


pic_22.jpg


В целом, все эти понятия в той или иной форме применимы к любой группе живых организмов в их естественной среде. Как внутри видов, так и между видами идет конкуренция за место обитания, за корм, за то чтобы стать или нет кормом и многое другое, что также относится к естественному отбору внутри и межвидовому. 


## Теория генетического алгоритма

Без названия.jpg

Как гласит Википедия - `Холланд Джон Генри` один из первых учёных, начавших изучать сложные системы и нелинейную науку - известен как отец `генетических алгоритмов`. 

`Генетические алгоритмы` стали особенно популярны благодаря работе Джона Холланда в начале 70-х годов и его книге `«Адаптация в естественных и искусственных системах»` (1975)

Разработанный Холландом генетический алгоритм включил в себя перечисленные выше понятия живой природы - особь, популяция, отбор, мутация, потомство, и в итоге стал мощным вычислительным инструментом.

Разберем теперь эти понятия с точки зрения вычислительных алгоритмов

Давайте представим себе, что перед нами электрощит с тумблерами в одну линию (для простоты). У каждого тумблера только два положения `вкл` и `выкл`, или с точки зрения программирования `1` и `0`. Каждый из тумблеров может включить или выключить какой-то прибор (сопротивление, диод, транзистор, конденсатор, цепь приборов..), который влияет, например, на напряжение в  определенном участке схемы. Причем эта цепь взаимосвязей такова, что просчитать по формулам "в лоб" и найти оптимальное решение слишком длительный процесс, как и прямой перебор всех возможных комбинаций, а он выражается формулой из комбинаторики  `размещения с повторением` k в степени n.

> И при 10 тумблерах это будет уже 5**10 = 9765625 вариантов!!

### Бот и популяция

Под популяцией в генетическом алгоритме понимается набор ботов (иногда их называют агентами).

image.png

В нашем случае, вид и содержание бота в точности соответствует тумблерам со значениями 0 и 1. В програмировании эту роль может выполнить список. 

Множество созданных ботов для задачи создают популяцию ботов.

Если мы применим наш бот к нашей электрической схеме с тумблерам и замерим, например напряжение в контрольном месте, то далее полученный результат мы уже сравним с нужным нам. 



### Отбор

image.png

Пусть напряжение должно достигнуть определенного значения - это и будет нашей целевой или оптимизационной функцией.

Для популяции мы создаем множество ботов в заданом количестве - `N`. И все они дадут различный результат.


image.png

По полученному результату, мы можем отсортировать ботов по их близости к нужному результату по абсолютному значению.

Теперь мы применим отбор, например, мы задаем число `N_suv` - количество отсортированых лучших ботов,  мы оставим, а остальные отбросим.

### Скрещивание и мутации

дерево.png

`N_surv` лкчших ботов мы отбираем в новую популяцию, которая должна быть также размера `N`. И нам необходимо сгенерировать новых ботов в количестве  `N - N_surv`.

image.png

Скрещивание можно организовать путем случайного выбора двух ботов-родителей из списка лучших и применить случайный выбор из значений их аналогичных генов.
-  если значения одинаковые, выбор будет из одного значения;
-  а если значения разные, то будет выбрано случайное значение гена из 2х значений.

Таким образом от двух родителей мы создадим нового бота

image.png

Как было ранее сказано, мутации (по сути ошибки копирования генов) в живой природе являются важнейшим механизмом изменчивости для адаптации к условиям окружающей среды. 

Мы также можем применить случайный мутации у случайного количества новых ботов в случайно выбранных генах, заменив их на противоположное. 

То есть даже там, где у обоих родителей было совпадение в гене, в результате случайной мутации может значение замениться на противоположное. И если эта мутация в результате применения функции проверки даст хороший результат, то мутировавший набор генов нового бота может дать новое поколение ботов

Далее, с помощью цикла `for in`  мы произадим `N - N_surn` новых ботов. И как как новые популяции ботов собраны, мы вновь можем применить функцию проверки и отобрать опять новое поколение лучших на новой эпохе эволюции. Подобный цикл эволюции можно также организовать через `for in`.

image.png

Тем самым через организацию циклов отборов ботов и рождения новых ботов новой популяции, мы организуем целенаправленный отбор.  Он намного эффективнее ищет лучших ботов, в нашем случае положения тумблеров (когда искомое значение через функцию проверки (в нашем случае замер напряжения и сравнение его с целевым) будет  максимально близко к целевому).

Нужно отметить, что чем сложнее система, то есть чем больше у нее степеней свобод, тем более широкое пространство выбора значений ботов. Как итог - множество различных путей поиска глобального минимума, где по дороге могут встречаться и локальные минимумы.

Все эти этапы и сложности очень напоминают работу нейронной сети. Хотя есть существенное отличие, в генетическом отборе нет взятие производной (частных производных) для направленного поиска.
