В задачах регрессии обязательно нужно обратить внимание на данные, которые необходимо предсказывать. Они могут сильно разнится и из-за этого качество обучения НС может сильно упасть. Например, те же зарплатные ожидания - кто-то запрашивает 20 тысяч, а кто-то 300 тысяч. Поэтому чаще всего используют нормирование данных.
* без нормирования данные могут быть в диапазоне от минус до плюс бесконечности, на примере зарплат от 0 и до бесконечности. При этом используют активационные функции **linear** и **relu**(если данные больше 0).
* при нормальном распределении с помощью **StandardScaler** диапазон данных после нормирования сложно предсказать, так как зависит от исходных данных, поэтому активационная функция только **linear**
* при min/max нормирование исходные данные приводятся к диапазону от 0 до 1 или от -1 до 1. Для примера если работаем с графиком уличной температуры, который колеблется от -30 до +40, то каждую отметку температуры делим на модуль самого большого числа, в данном случае 40. В данном случае можем использовать почти любые активационные функции, в соответствии с диапазоном нормирования.

Выбор активационой функции является очень важным элементом подбора архитектуры, так как ошибка может привести к невозможности решить задачу. Например, сделали нормальное распределение с диапазоном от -3,2 до +17,5 и взяли активацию **sigmoid()**. В результате этого нейронка на выходе будет давать числа в диапазоне от 0 до 1.

Важность подготовки данных для НС также очень высока, рассмотрим этот вопрос далее.

Рассмотрим на примере уже известной таблицы резюме. Все данные в ней должны быть оцифрованы. Данные можно разделить на несколько блоков:
* числовые данные - зарплатные ожидания
* категориальные данные - все оставшиеся данные представлены в виде текста, но это не один блок. Просмотрим столбюец занятость и увидим, что в этом столбце всего 4 значения:
  * полная занятость
  * частичная занятость
  * стажировка
  * проектная работа
 
  у пользователя может быть несколько из данных вариантов, но это не меняет того факта, что позиций всего 4. Поэтому эти данные не логично обрабатывать как текстовые. Поэтому каждый вариант будем кодировать вектором состоящим из 4 элементов [0, 0, 0, 0]. Те значения которые есть у пользователя в этом векторе будет ставиться 1. например для пользователя с полной и частичной занятостью вектор будет выглядеть так [1, 1, 0, 0]
* текстовые данные - искомая должность, предыдущая должность, текстовое описание опыта работы.

В данном примере есть вариативность со столбцом возвраст. С одной стороны можно данные подавать как есть, с другой стороны человек 27 и 28 лет не стльно отличаются, поэтому можно разбить на группы возврастов и в НС подавать вектор к какой группе возвраста относится пользователь.

С городами применили такую же логику и создали 4 группы:

Отметим, что это не единственно правильный вариант, например города можно разделить по федеральным округам и тогда групп будет больше.
