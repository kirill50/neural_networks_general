Что же такое звук?

Презентация Занятие 11. Обработка аудиосигналов с помощью нейронных сетей (1).jpg

Звук – это физическое явление, которое можно описать как **_распространение механических колебаний в виде волн_** в некоторой среде. Среда распространения может быть твердой, жидкой или газообразной. Скорость распространения волн зависит от плотности среды и может сильно отличаться в разных средах. Звуки, которые слышит человек, распространяются в воздухе; скорость их распространения зависит от температуры, влажности, ветра и других факторов. Приблизительная скорость распространения звука в воздухе составляет **340** метров в секунду. 

Представьте себе волны, которые появляются и движутся на поверхности воды, если в нее бросить камень. Отличие распространения звуковых волн состоит в том, что на воде вы видите распространение волн в двумерной среде, а звук распространяется в трехмерном пространстве.

одинаковых букв можно заметить общие черты даже на глаз.

Коснемся важнейших свойств звука. Как и любая волна, звук характеризуется **_амплитудой_** и **_частотой_**.

**_Амплитуда_** (amplitude) характеризует звуковое давление и громкость звука. Чем больше амплитуда, тем громче звук, и наоборот. Громкость принято измерять в **_децибелах_** (dB).

**_Частота_** (frequency) определяется как количество периодов колебания звуковой волны за одну секунду и измеряется в **_герцах_** (Hz). На слух частота характеризует высоту тона. Чем выше частота, тем больше периодов колебания делает волна. Соответственно, чем выше звук, тем больше колебаний, и наоборот.
 
В природе звуки, как правило, сложные, то есть состоят из одновременно звучащих волн разной частоты. Поэтому часто говорят не об одной частоте, а о так называемом **_спектре_** волны - наборе составляющих звук элементарных волн разных частот (**_гармоник_**). Структура спектра аудиосигнала определяет его тембровую окраску и звуковысотное положение.

Природные возможности человека в восприятии звука имеют ограничения. На иллюстрации показана область звуковых частот и громкости, доступная для восприятия человеком. В ее пределах располагаются зоны речи и музыки.

Согласно исследованиям слуха, человек воспринимает звуки в диапазоне частот примерно от **20** Hz до **20000** Hz (**20** kHz). Звуки выше и ниже границ этого диапазона, конечно, существуют, но человек их не слышит, хотя и может на них реагировать (например, на сверхнизкие частоты).

Если громкость звука превышает **130** dB (_"болевой порог"_, соответствует шуму взлетающего вблизи самолета), человек испытывает боль и слух уже повреждается, а громкость в **180** dB смертельна, что применяется в шумовом оружии.

Звук частотой ниже границы слышимости (**20** Hz) называют **_инфразвуком_**, выше (от **20** kHz до **1** GHz) – **_ультразвуком_**, от **1** GHz — **_гиперзвуком_**.

В приложении к нашему курсу необходимо разобраться, как звук преобразуется в цифровую форму.

Звуковое давление в каждый момент времени имеет какое-то значение. Для оцифровки звука можно измерять это значение через одинаковый шаг времени. В результате получится набор отсчетов как дискретное (числовое) представление звуковой волны. Но как часто нужно делать отсчеты, чтобы избежать потерь? Ответ дает теорема Котельникова!

Смысл теоремы в том, что **_частота дискретизации (измерения значений звукового давления) должна быть не менее удвоенной максимальной частоты аналогового сигнала_**.

Например, если необходимо оцифровать звук, захватывая все составляющие его частоты от **0** до **15** kHz, то частота дискретизации должна быть не менее **30** kHz, то есть в **1** секунду нужно сделать не менее **30000** отсчетов. Только в этом случае требуемый набор частот будет  полностью представлен в оцифрованном звуке. Более высокие частоты будут искажены или подавлены.

Хорошо, что все это не нужно делать вручную; есть набор библиотек с необходимыми инструментами.

Существует множество инструментальных средств для работы со звуком. Среди них:

1. **PyAudio** – библиотека визуализации звуковой информации в ноутбуке.
2. **SpeechRecognition** – инструменты для распознавания речи (преобразования речи в текст).
3. **sTTS** – инструменты для преобразования текста в речь. Решают обратную для **SpeechRecognition** задачу.
4. **LibROSA** – инструменты для извлечения значимых признаков из произвольного звука. Основная библиотека, которую вы будете использовать в анализе звука.

**_Спектрограмма_** – это визуальный способ представления уровня громкости, или энергии, сигнала во времени на различных частотах, составляющих вместе сложную волну. Часто изображается в виде тепловой карты.

Спектрограмма получается следующим образом:

Звук раскладывают по частотам и выделяют, на каких частотах энергия (громкость) выше. На картинке представлен результат таких преобразований. До двенадцатой секунды можно предположить, что звук отсутствует или очень тихий, затем присутствуют звуки различной высоты и громкости с промежутками тишины.

Спектрограмма отображается с помощью функции:

`librosa.display.specshow()`

**_Частота (или скорость) пересечения нуля_** соответствует общей оценке тембра звучания по 
шкале высокий/низкий, глухой/звонкий, шипящий/свистящий. По графику можно быстро определить, где низкий звук, а где высокий, поскольку у высокого звука пересечений нуля больше, чем у низкого.

**_Мел-частотные кепстральные коэффициенты_** представляют собой небольшой набор признаков (обычно **10–20**), которые кратко описывают структуру спектра звука. Фильтры, которыми выделяются коэффициенты, основаны на особенностях человеческого восприятия звука: древним людям важнее было распознавать низкочастотные звуки, что отразилось в эволюции органов слуха. Странное название дали авторы методики: математически преобразованный спектр назвали "кепстром".

Кепстральные коэффициенты по скользящему окну в аудиосигнале вычисляют с помощью функции:

`librosa.feature.mfcc()`

**_Хромаграмма_** или **_хроматический вектор_** обычно представлен набором признаков из **12** элементов, характеризующих интенсивность полутонов музыкальной гаммы (C = до, C# = до диез, D = ре, D# = ре диез и так далее) в исследуемом звуке. Используется для выделения и анализа мелодических признаков в музыкальном аудиосигнале.

Хромаграмма вычисляется с помощью функции:
