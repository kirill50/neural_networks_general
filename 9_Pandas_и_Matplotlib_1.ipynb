## Повторение пройденного материала (1 занятие)

Освежите память и повторите уже пройденный материал.

Для начала подключите модуль pyplot библиотеки matplotlib. Задайте псевдоним plt:

````
# Испорт модуля pyplot библиотеки matplotlib под псевдонимом plt
import matplotlib.pyplot as plt

# Команда обеспечивает вывод графиков в Google Colaboratory
%matplotlib inline
````

Для визуализации графиков вам понадобится массив. Создайте его с помощью библиотеки NumPy:

````
import numpy as np                        # Подключение библиотеки numpy
x = np.linspace(0, 15, 101)               # Генерация последовательности из 101 числа, дающих
                                          # равномерное разбиение интервала [0..15]
print(x.shape)                            # Вывод размерности сгенерированного массива
print(x)                                  # Вывод самого массива
````

Объект класса plt – это ваше полотно. 

Применяя к нему различные методы, вы наносите на полотно графики, надписи, текст, элементы стиля и т.п. Для начала отрисуйте график массива `x`:

````
plt.plot(x)                               # Рисование графика
plt.show()                                # Фиксация (вывод) графика
````

В переменную `y` запишите `cos(x)` и отрисуйте график:

````
y = np.cos(x)                             # Вычисление значений функции косинус на сетке значений x
plt.plot(y)                               # Рисование графика
plt.show()                                # Фиксация графика
````

Для создания более красивых и наглядных графиков полезно знать различные способы их оформления. Вот они:

* подписать оси
```
plt.xlabel('название оси X')
plt.ylabel('название оси Y')
```

* подписать название графика
```
plt.title('название графика')
```

* отобразить легенду графиков (легенда - это параметр **label** метода `.plot()`)
```
plt.legend() 
```

* изменить размер полотна
```
plt.figure(figsize=(12, 4))
```

* отрисовать сетку
```
plt.grid()
```

* изменить стиль графика
```
plt.plot(x, z, 'c--')
```

````
plt.figure(figsize=(12, 4))               # Указание размера полотна в дюймах

z = np.sin(x) / 2                         # Вычисление значений функции sin(x)/2
plt.plot(x, y, label='cos(x)', c='b')     # Рисование графика значений y, указание названия и цвета
plt.plot(x, z, label='sin(x)/2', c='g')   # Рисование графика значения z, указание названия и цвета

plt.title('sin(x)/2 и cos(x)')            # Указание заголовка графика

plt.xlabel('ось X')                       # Указание подписи оси х
plt.ylabel('Y, Z')                        # Указание подписи оси y

plt.grid()                                # Отрисовка сетки на полотне
plt.legend(loc='upper left')              # Отображение легенды на графике в левом верхнем углу
plt.show()                                # Фиксация графика
````

Параметр `loc=` метода `.legend()` определяет местоположение легенды на графике. В данном случае легенда расположена в левом верхнем углу. Если не указать этот параметр, то местоположение подберется автоматически таким образом, чтобы закрывать наименьшую часть графика.

Рассмотрите различные стили графиков:

````
plt.figure(figsize=(15, 1))
plt.plot(x, z, 'c--')                     # Прерывистая голубая линия
plt.show()

plt.figure(figsize=(15, 1))
plt.plot(x, z, 'y*')                      # Желтые звездочки
plt.show()

plt.figure(figsize=(15, 1))
plt.plot(x, z, 'gs')                      # Зеленые квадраты
plt.show()

plt.figure(figsize=(15, 1))
plt.plot(x, z, 'c^')                      # Голубые треугольники
plt.show()

plt.figure(figsize=(15, 1))
plt.plot(x, z, 'ro')                      # Красные кружки
plt.show()
````

В метод plot можно передавать еще много различных параметров, например:
- **label** - описание графика
- **color** - цвет линии
- **linewidth** - ширина линии
- **marker** - маркер
- **markersize** - размер маркера
- **mec** - цвет маркера
- и другие

Полное описание есть в документации библиотеки.
Попробуйте создать уникальный график, используя эти параметры:


````
plt.figure(figsize=(14, 7))               # Указание размера полотна в дюймах

plt.plot(x, y, label = "sin(x)", color='#4b0082', linewidth=4,
         marker='h', markerfacecolor='lightgreen', markeredgewidth=2,
         markersize=10, markevery=5)

plt.plot(x, np.cos(x) / 2, label = 'cos(x)/2', linewidth=3, color='lightcoral',
         marker='D', markeredgecolor='black', markevery=5)

plt.title("y = sin(x)")                   # Заголовок графика
plt.xlabel('Переменная X')                # Подпись оси x
plt.ylabel('Переменная Y')                # Подпись оси y

plt.legend()                              # Отрисовка названий графиков ("легенды")
plt.grid()                                # Отрисовка сетки
plt.show()                                # Фиксация графика
````

## Параметры linestyle и marker

Посмотрим два параметра для оформления линий графиков.

###Параметр **linestyle**

Позволяет менять стиль линии.

Имеет следующие доступные варианты:
* `'-'`, `'solid'`
* `'--'`, `'dashed'`
* `'-.'`, `'dashdot'`
* `':'`, `'dotted'`
* `'steps'`
* `'None'`, `' '`,  `''`

Рассмотрите каждый из доступных вариантов.

````
line_styles = ['-', '--', '-.', ':', 'steps', 'None']  # Создание списка с разными стилями линий
for ls in line_styles:                                 # Цикл для последовательного отображения всех вариантов
    plt.figure(figsize=(15, 6))
    plt.plot(x, z, c='r', ls=ls)                  # Подстановка значений из списка в параметр ls=
    
    plt.title('0.5 sin(x)')
    plt.xlabel('х')
    plt.ylabel('y')
    plt.show()
````

Обратите внимание, что последний график не пустой. Вы сможете убедиться в этом, если включите отображение маркеров. Для этого понадобится параметр **marker**:

###Параметр **marker**

Позволяет менять стиль маркеров.

Имеет следующие доступные варианты:

> `'.'` `','` `'o'` `'s'` `'p'` `'*'` `'h'` `'H'` `'x'` `'d'` `'D'` `'_'` `'|'` `None`

> `'v'` `'^'` `'<'` `'>'` аналогичны `'1'` `'2'` `'3'` `'4'`

Рассмотрите некоторые из них на примере.

````
marker_styles = ['+', 'o', 'D', '1', 'None']       # Создание списка с разными стилями маркеров

for marker in marker_styles:                       # Цикл для последовательного отображения всех вариантов
    plt.figure(figsize=(15, 6))
    plt.plot(x, z, c='r', marker=marker, mec='b')  # Параметр mec задает цвет маркера,
                                                   # а параметр marker - тип маркера
    plt.title('0.5 sin(x)')
    plt.xlabel('х')
    plt.ylabel('y')
    plt.show()
````

Теперь убедитесь, что график со стилем линии **None** действительно не пустой. Для этого параметру **linestyle** задайте значение **None**, а параметру **marker** задайте любое значение, кроме **None**:

````
plt.figure(figsize=(15, 6))
plt.plot(x, z, c='r', ls='None', marker='o')

plt.title('0.5 sin(x)')
plt.xlabel('х')
plt.ylabel('y')
plt.show()
````

## Вывод нескольких графиков (2 способа)

Иногда необходимо вывести сразу несколько графиков, чтобы было удобно их сравнить. Это можно сделать двумя способами.

### Метод `.subplot()`

Используя метод `.subplot()`, вы создаете таблицу, в ячейки которой помещаете графики. 

Для начала создайте таблицу из одного столбца и двух строк для вывода двух графиков:

В метод `.subplot()` подается три числа:
- количество «строк» графиков;
- количество «столбцов» графиков;
- порядковый номер графика, для которого задаете параметры.

Исходя из условий, команды будут выглядеть следующим образом:

        plt.subplot(211) # – для первого графика;
        plt.subplot(212) # – для второго графика.
        
Перенесите команды в ячейку кода, чтобы проверить результат:


````
plt.figure(1, figsize=(10, 5))

# 2 - количество строк, 1 - количество столбцов, 1 - порядковый номер графика
plt.subplot(211)
plt.title('Два графика с общей осью х')
plt.plot(x, y, 'b-.', label='cos(x)')
plt.legend()

# 2 - количество строк, 1 - количество столбцов, 2 - порядковый номер графика
plt.subplot(212)
plt.plot(x, z, 'r--', label='0.5 sin(x)')
plt.xlabel('общая ось х')
plt.legend()

plt.show()
````

Отлично! А теперь попробуйте создать таблицу 2 на 2, но вывести не четыре, а три графика.

Ячейки нумеруются слева-направо, то есть следующим образом:

```
1 | 2
3 | 4
```

````
plt.figure(1, figsize=(10, 8))

plt.subplot(221) # Таблица 2 на 2, работаем с графиком в ячейке 1
plt.title('Вывод трех графиков')
plt.plot(x, y, 'b-.', label='cos(x)')
plt.legend()

plt.subplot(222) # Таблица 2 на 2, работаем с графиков в ячейке 2
plt.plot(x, z, 'g--', label='0.5 sin(x)')
plt.xlabel('общая ось х')
plt.legend()

plt.subplot(224) # Таблица 2 на 2, работаем с графиком в ячейке 4
plt.plot(x, z, 'r--', label='0.5 sin(x)')
plt.xlabel('общая ось х')
plt.legend()

plt.show()
````

Как видите, в третьей ячейке нет графика, так как вы его не указали.


---


**Обратите внимание**, что с графиками в ячейках можно работать как с обычными: указывать стиль линий, маркеров, отображение легенды и так далее.


---


Размер полотна в данном случае определяет, сколько пространства будет выделено под все графики вместе, а не под каждый в отдельности.


### Метод `.subplots()`

```
figure, axes = plt.subplots(2, 2, sharex=False, sharey=False)
```

Метод `.subplots()` тоже создает таблицу, которую вы заполните графиками. Поэтому сначала подайте два числа – количество строк и столбцов.


Параметры `sharex` и `sharey` могут принимать значения **False** либо **True**. Если **False**, то у каждого графика будет отображена своя ось x/y. Если **True**, то ось x/y будет общей для всех графиков.

`.subplots()` возвращает значения, которые мы передаем в переменные **figure** и **axes**.


---


```
axes[0, 0].plot(x, y, label='sin(x^2)', c='r')
axes[0, 0].legend()
```

**Figure** - это контейнер самого верхнего уровня, в котором могут находиться контейнеры **axes**.

С помощью **axes[]** вы можете обращаться к разным контейнерам. Только теперь вы делаете это не по порядковому номеру, а по индексу строки и столбца.

Заполните все ячейки и проверьте результат:

````
# Создание данных для рисования графика
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

# Создание пространства для подграфиков
figure, axes = plt.subplots(2, 2, sharex=False, sharey=False, figsize=(12, 8)) 

# График с координатами положения (0, 0) на полотне
axes[0, 0].plot(x, y, label='y=sin(x^2)', c='r')
axes[0, 0].legend() 

# График с координатами положения (0, 1) на полотне
axes[0, 1].plot(x, np.abs(y), label='np.abs(y)', c='b')
axes[0, 1].legend()

# График с координатами положения (1, 0) на полотне
axes[1, 0].plot(x, y ** 3, label='y**3', c='g')
axes[1, 0].legend()

# График с координатами положения (1, 1) на полотне
axes[1, 1].plot(x, np.arccos(y), label='arccos(y)', c='y')
axes[1, 1].legend()
plt.show()
````

Есть один большой контейнер **figure**, а в нем четыре контейнера **axes** с графиками.

## Столбчатая диаграмма

Создайте столбчатую диаграмму. Для этого используйте метод `.bar()`:


```
plt.bar(x, height, align='center', color='r', linewidth=20)
```

- **x** и **height** - это две переменные для осей x и y соответственно
- `align='center'` - способ выравнивания столбцов по оси x. Может принимать значения "center" (по центру) и "edge" (по краю)
- `color='r'` - цвет столбцов графика
- `linewidth=20` - ширина столбцов


---


Метод `.xticks()` позволяет задать столбцам имена
```
plt.xticks(x, ['Январь', 'Февраль', 'Март', 'Апрель', 'Май'])
```
- **x** - переменная, в которой хранятся данные для оси x
- [...] - массив значений для каждого столбца

А теперь проверьте, как это работает на практике:

````
# Данные для оси x
x = np.arange(5)
# Данные для оси y
height = [20, 23, 25, 24, 26]

# Построение столбчатой диаграммы
plt.bar(x, height, align='center', color='r', linewidth=20)

plt.title('Продажи товара Х по месяцам')

# Подпись оси y
plt.ylabel('млн руб.')                    
# Метки шкалы оси x
plt.xticks(x, ['Январь', 'Февраль', 'Март', 'Апрель', 'Май'])
plt.show()
````

Посмотрите, что будет, если изменить способ выравнивания столбцов на edge:

````
x = np.arange(5)
height = [20, 23, 25, 24, 26]

# Построение столбчатой диаграммы, используем выравнивание по краю
plt.bar(x, height, align='edge', color='b', linewidth=20)

plt.title('Продажи товара Х по месяцам')

plt.ylabel('млн руб.')                    
plt.xticks(x, ['Январь', 'Февраль', 'Март', 'Апрель', 'Май'])
plt.show()
````

Как видите, теперь столбец располагается справа от отметки на оси, а не по центру.

## Гистограмма

Теперь создайте гистограмму с помощью метода `.hist()`:


```
plt.hist(x, bins=10, color='b', linewidth=10)
```

- **x** - переменная с массивом значений
- `bins=10` - количество подгрупп, в которые объединятся данные
- `color='b'` - цвет гистограммы
- `linewidth=10` - ширина столбцов гистограммы

````
# Создание массива чисел, распределенных по нормальному закону
x = np.random.normal(0, 1, 10000) 

# Построение гистограммы
plt.hist(x, bins=10, color='b', linewidth=10)

plt.title('Гистограмма распределения значений в массиве х')
plt.show()
````

Чтобы лучше понять за что именно отвечает `bins`, увеличьте значение в 10 раз:

````
x = np.random.normal(0, 1, 10000) 

# Построение гистограммы
plt.hist(x, bins=100, color='b', linewidth=10)

plt.title('Гистограмма распределения значений в массиве х')
plt.show()
````

Создайте переменные **x** и **y** со случайными значениями, а затем, используя метод `.scatter()`, отобразите точки на графике:

## Тепловая карта (Heatmap)

Иногда появляется необходимость в отображении данных, которые зависят от двух переменных. Это можно сделать, используя тепловую карту.

Для работы с тепловыми картами создайте массив, состоящий из нулей и единиц:


````
# Генерация данных для тепловой карты
heatmap_array = np.zeros((9, 9))          # Создание массива из нулей
heatmap_array[4] = 1                      # Замена строки нулей на единицы
heatmap_array[:, 4] = 1                   # Замена столбца нулей на единицы
print(heatmap_array)
````

Теперь с помощью метода `.imshow()` выведите тепловую карту созданного массива:

````
plt.imshow(heatmap_array)                 # Вывод массива по цветам палитры
plt.show()
````

Все единицы окрасились в желтый, а нули – в фиолетовый. Чтобы увидеть всю палитру от нуля до единицы, используйте метод `.colorbar()`:


````
plt.imshow(heatmap_array)                 # Вывод массива по цветам палитры
plt.colorbar()                            # Цветовая полоса colorbar
plt.show()
````

Вы можете менять используемую палитру с помощью параметра `cmap`. 

Рассмотрите несколько примеров:

````
plt.imshow(heatmap_array, cmap='hot')     # cmap задает цветовую палитру отображения
plt.colorbar()
plt.show()
````

````
rnd = np.random.random((9, 9))            # Массив из рандомных значений от 0 до 1
plt.imshow(rnd, cmap='bone')              # cmap задает цветовую палитру отображения
plt.colorbar()
plt.show()
````

````
plt.imshow(rnd, cmap='gray')              # cmap задает цветовую палитру отображения
plt.colorbar()
plt.show()
````

Есть и другие цветовые палитры. Подробнее о них можно узнать из [документации библиотеки **Matplotlib**](https://matplotlib.org/).
