## Бесплатная библиотека TTS

Начнем с ее установки:

Данная библиотека находится в свободном доступе, использование бесплатно.
Документация по ссылке: https://gtts.readthedocs.io/en/latest/

```python
# Импорт библиотеки google-cloud-texttospeech
import gtts 

# Основной интерфейс для перевода текста в аудио
from gtts import gTTS 

# Используем для проигрывания аудио
from IPython.display import Audio 
```

Словарь со всеми доступными языками. К сожалению, для русского языка доступен только один голос:

```python
gtts.lang.tts_langs()
```

```python
# Запишем стихотворение Блока в переменную

text = """Ночь, улица, фонарь, аптека,
          Бессмысленный и тусклый свет.
          Живи ещё хоть четверть века —
          Всё будет так. Исхода нет.

          Умрёшь — начнёшь опять сначала
          И повторится всё, как встарь:
          Ночь, ледяная рябь канала,
          Аптека, улица, фонарь."""

# Преобразовываем текст в аудио
tts = gTTS(text, lang='ru') 

# Сохраняем в формате мр3
tts.save('blok.mp3')        
```

```python
# Прослушаем полученное аудио
Audio('blok.mp3')
```

Результат преобразование текста в речь получился неплохой. Интонация присутствует, но сказать, что это произносит человек вряд ли можно.

###Препроцессинг

Данная библиотека позволяет изменять параметры обработки текста.


```python
# Преобразование текста в речь без препроцессинга

tts = gTTS('РФ - страна с огромной территорией', lang='ru') 
tts.save('rf.mp3')
Audio('rf.mp3')
```

Препроцессинг позволяет модифицировать входной текст под конкретные нужды. Например, если в тексте присутствуют аббревиатуры и при произношении их нужно расшифровывать, то можно указать это, добавив в токенайзер соответствующие аббревиатуры и их расшифровку.

```python
# Импорт препроцессоров
from gtts.tokenizer import pre_processors 

# Добавляем замену 
gtts.tokenizer.symbols.SUB_PAIRS.append(('РФ', 'Российская федерация')) 

# Пишем текст
text = 'РФ - страна с огромной территорией' 

# Синтезируем речь и сохраняем для прослушивания
tts = gTTS(text, lang='ru') 
tts.save('rf.mp3')
Audio('rf.mp3')
```

Теперь все аббревиатуры "РФ", которые встретятся в тексте будут расшифровываться.

Можно сделать больше замен (но лучше не надо).

```python
# Добавляем в список замен дополнительные элементы (строка, ее замена)

gtts.tokenizer.symbols.SUB_PAIRS.append(('стр.', 'cтрана'))                        
gtts.tokenizer.symbols.SUB_PAIRS.append(('огр.', 'огромной'))
gtts.tokenizer.symbols.SUB_PAIRS.append(('терр.', 'территория'))
gtts.tokenizer.symbols.SUB_PAIRS.append(('дейст.', 'действительно'))
gtts.tokenizer.symbols.SUB_PAIRS.append(('бол.', 'большая'))
```

```python
# Плохочитаемый текст, но благодаря заменам, аудио будет звучать нормально
text = 'РФ - стр. с огр. терр. Ее терр. дейст. бол.' 

# Синтезируем речь и сохраняем для прослушивания
tts = gTTS(text, lang='ru') 
tts.save('1.mp3')
Audio('1.mp3')
```

Как правило, препроцессинг используется для замены часто встречающихся аббривиатур и сокращений (стр., РФ, ред. и прочее), а также для задания стиля прочтения пунктуации (Полное игнорирование, игнорирование запятых и т.д.)

На этом знакомство с бесплатной библиотекой для синтеза речи от google окончено.
