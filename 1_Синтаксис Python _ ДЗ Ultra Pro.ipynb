### Задача 1

Вы отправляетесь в отпуск в Америку. У себя на родине вы привыкли все считать в рублях, километрах и литрах.

Приехав в Америку, вы запланировали недельную поездку от Нью-Йорка до Лос-Анджелеса на машине. Весь день вы будете ехать, а ночью останавливаться в отеле.

Исходые данные:
* Расстояние: **2790** миль;
* Стоимость аренды авто на неделю: **820** долларов;
* Стоимость одной ночи в отеле (всего **6** ночей): **110** долларов;
* Расход бензина: **15.8** миль на один галлон;
* Стоимость одного галлона **3.89** доллара;
* Курс доллара **77.4** руб.

Необходимо вычислить в рублях, сколько потребуется денег, чтобы проехать этот путь на машине.

При выводе на экран все значения округлите до двух знаков после запятой.

```python
# Ваше решение
class Budgetting:

  def __init__(self, distance: int, car_rental :int,
               accommodation :int, duration :int,
               fuel_consumption,
               gallon_price, exchange_rate):
        self.distance = distance
        self.rental = car_rental
        self.accommodation = accommodation
        self.duration = duration
        self.fuel_consumption = fuel_consumption
        self.fuel_price = gallon_price
        self.rate = exchange_rate

  def car_expenses(self):
        fuel_expenditures = self.distance / self.fuel_consumption *\
                            self.fuel_price
        total = fuel_expenditures + self.rental
        return total * self.rate 

  def accomodation_expenses(self):
        total = self.accommodation * self.duration
        return total * self.rate

  def grand_total(self):
        total = self.car_expenses() + self.accomodation_expenses()
        total = round(total, 2)
        return f'Total expenses for the jorney will be {total:,} rur'
```

```python
budget = Budgetting(2790, 820, 110,
                    6, 15.8, 3.89, 77.4)
budget.grand_total()
```

### Задача 3

Попробуйте написать простую карточную игру. 
Играют два игрока - **player_1** и **player_2**.
У каждого из них по **6** карт.

* **player_1 = [6, 10, 7, 8, 9, 6]**
* **player_2 = [10, 7, 9, 6, 10, 7]**

Необходимо, используя операторы `for`, `if` и `elif`, сыграть одну партию. 
Игроки ходят по очереди, и тот, у кого достоинство карты больше, забирает себе карту противника. Сыгранные карты больше не используются. Достоинства карт в списке берутся по порядку, например:

 * первый ход: **(6, 10)**;
 * второй ход: **(10, 7)**.

Побеждает тот, у кого по окончании игры на руках меньшая сумма достоинств карт.

Кто победил, **player_1** или **player_2**?

```python
# Ваше решение
player_1 = [6, 10, 7, 8, 9, 6]
player_2 = [10, 7, 9, 6, 10, 7]

player_1_won = []
player_2_won = []

for turn_1, turn_2 in zip(player_1, player_2):
    if turn_1 < turn_2:
        player_2_won.append(turn_1)
    elif turn_1 > turn_2:
        player_1_won.append(turn_2)
    else: 
        print('Unexpolered case')
    
    player_1_result = sum(player_1_won)
    player_2_result = sum(player_2_won)

if player_1_result < player_2_result:
    print(f'Player_2 is the winner!')
elif player_1_result > player_2_result:
    print(f'Player_1 is the winner!')
else:
    print(f"Looks like that it's draw")
```

### Задача 4

Введите три слова через пробел. Преобразуйте их таким образом, чтобы все буквы стали заглавными и между буквами в каждом слове появились дефисы.

Например: Вставай на лыжи --> В-С-Т-А-В-А-Й Н-А Л-Ы-Ж-И

```python
# Ваше решение
def proccess_words(several_words :str):
    final_string = ''
    words_list = several_words.split(" ")

    for word in words_list:
        word = word.upper()
        string = '-'.join(word)
        final_string = final_string + ' ' + string
    return final_string
```

```python
proccess_words('Вставай на лыжи')
```

### Задача 5

Есть план этажа здания. На этом этаже 4 комнаты (1, 2, 3, 4), все выходят в общий коридор. Комнаты не сообщаются между собой: чтобы перейти из одной комнаты в любую другую, нужно пройти через коридор.

image.png

Создайте функцию, которая будет проверять, возможен ли переданный ей путь по комнатам. Коридор будет обозначаться буквой «H» (Hall).

* Путь может начинаться или кончаться в коридоре.
* В качестве инпута будут передаваться только цифры 1-4 и/или буква «H».
* Комнаты не будут повторяться.

```python
# Ваше решение
def right_route(route :list):
    if ((route[0] != 'H') or (route[-1] != 'H')):
        return('Маршрут должен начинаться или заканчиваться в коридоре!')
    else:
        #possible destinations
        possible_rooms = [1,2,3,4]
        for i, location in enumerate(route):

            #every odd location should be the 'H'
            if i%2 == 0:
                if location != 'H':
                    return('Маршрут неверный!')

            #every even location should be the room number
            #every room can be visited just once
            if i%2 != 0:
                if location in possible_rooms:
                    possible_rooms.remove(location)
                else:
                    return('Маршрут неверный!')
                 
        return('Поздравляю маршрут пройден успешно!')
```

```python
right_route(['H', 1, 'H', 4, 'H'])
```
