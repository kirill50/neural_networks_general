## Синтез речи

рпо.jpg

**Синтез речи или Text-to-Speech (TTS) — технология преобразования текста в речь.**

Работает синтез речи следующим образом: человек набирает какой-либо текст и отправляет его в модель TTS, а она, в свою очередь, генерирует на выходе аудиодорожку, в которой этот текст произносится голосом. Вся магия происходит внутри модели TTS.

На сегодняшний день, существует 2 основных подхода к созданию моделей TTS:

1.   Concatenative (конкатенативный)
2.   Parametric (параметрический)



Скриншот 03-11-2021 122038.jpg

Конкатенативный подход имеет не сложную архитектуру. Самое главное, нужно собрать базу речевых единиц (например слов или фонем). База должна насчитывать огромное количество речевых единиц, произнесенных с разной скоростью, интонацией и т.д. На каждую речевую единицу в базе должен быть соответствующий отрезок аудио.

Имея базу речевых единиц, подаем текст и вытаскиваем из базы соответствующие фрагменты аудио, которые затем склеиваются и сглаживаются для более плавного произношения.

Основные минусы при таком подходе можно выделить следующие:

*   Необходима большая обучающая база
*   Имеет характерные артефакты склейки
*   Синтезируемая речь монотонна
*   Не может генерировать звук, не встречающийся 
в обучающей выборке

Плюсом такого подхода является быстрая скорость работы.





Скриншот 03-11-2021 124359.jpg

Большинство современных систем TTS построены на основе параметрического подхода. Эти системы намного более сложные (некоторые параметрические системы TTS насчитывают в себе до 5 нейронных сетей). Параметрические TTS генерируют аудио не на основе базы речевых единиц, а на основе параметров, таких как F0, длина фонемы, высота звука и т.д. Нейронные сети обучаются задаче предсказания этих самых параметров (на основе поданного текста).
Поэтому нужна обучающая база, которая содержит текст и его аудио (но база намного меньше, чем при конкатенативном подходе).

Минусы такой системы в основном связаны с вычислительными мощностями и, соответственно, с низкой скоростью работы, но самые современные модели работают уже довольно быстро.

Что касается плюсов, то это:

*   Естественная и плавная генерируемая речь

*   Разнообразие интонаций

*   Требует меньшего объема данных для обучения



## Область примененния

Область применения связана с любой задачей, где нужно озвучивать текст.

*   Голосовой помощник. Настройка голосового робота для того, чтобы он отвечал на запросы клиентов без привлечения живых операторов.

*   Озвучивание контента. С помощью синтеза речи можно озвучивать книги, SMS-сообщения, документы и т.д. любым голосом.

*   Скрининг соискателей. Синтез речи можно использовать для автоматизации массового найма сотрудников. С помощью голоса можно проводить собеседования и отбирать кандидатов

*   И другие.

## Библиотеки для синтеза речи

Систем для синтеза речи на сегодняшний день становится всё больше и больше. Многие крупные компании разрабатывают свои системы, со своими фишками. Вот некоторые библиотеки TTS, которые поддерживают русский язык:

*   Google Text-to-Speech - синтезатор речи от Google. Эта библиотека умеет преобразовывать строку в mp3 файл с речью. Т.к. за этой библиотекой стоит Google, то на выбор имеется много языков, включая русский. Бесплатная.

*   Google-cloud-texttospeech – платная библиотека от Google. Она имеет широкий набор языков и голосов, и есть голоса, созданные с помощью Wavenet.

*   QtSpeech - это кроссплатформенная библиотека, основанная на Qt, которая предоставляет общий кроссплатформенный API для доступа и использования системных механизмов TTS.

*   Tinkoff Voice Kit - платная библиотека с системами TTS и STT. Работает через API.

*   Yandex SpeechKit - платная библиотека с системами TTS, STT и многое другое. Работает через API.
