Используя генетический алгоритм, найдите корни уравнения:

$f(xy)=7x^2 - y + \frac{y^3 - x^2}{3x} + y^2 - 21x$

*P.s. постарайтесь по максимуму использовать возможности библиотеки numpy для ускорения процеесса вычислений*

**Решение**

```python
# Подключение библиотек
import numpy as np
```

```python
# Задание рабочей функции
def f(x, y):
    return 7*x**2 - y + (y**3 - x**2) / (3*x) + y**2 - 21*x
```

```python
# Проверка работы функции для двух случайных значений
x = np.random.random()
y = np.random.random()
print(f'x={x}')
print(f'y={y}')
print(f'f(xy)={f(x,y)}')
```

Будем решать задачу, используя возможности библиотеки numpy

Боты будут аналогичны тем, что использовались в учебном ноутбуке

```python
# Задание длины бота
botlength = 8

# Задание размера популяции
populationlength = 100

# Задание стартового значения мутации
mutation = 0.6

# Задание количества выживших
nsurv = 20

# Задаем количество новых ботов
nnew = populationlength - nsurv

# Задание количества эпох
epochs = 1000
```

Создадим python-функцию расчета значения указанной в задании функии, на основе значений ботов

```python
def fbybot(bot):
    # Расчет значения x
    x = bot[0] + 5*bot[1] + 10*bot[2] + 25*bot[3]

    # Расчет значения y
    y = bot[4] + 5*bot[5] + 10*bot[6] + 25*bot[7]

    # Возвращаем значение функции
    return f(x, y)
```

Зададим стандартный набор функций учебного ноутбука, но с использованием numpy

```python
def getsurvpopul(val):
    """ Функция получения лучших ботов 
        val - массив ошибок ботов    
    """
    
    # Получаем отсортированный список ошибок
    sortindex = np.abs(val).argsort()

    # Возвращаем лучшие боты и их значения ошибок
    return population[sortindex].copy(), val[sortindex]
```

```python
# Создание популяции
population = np.random.uniform(-10, 10, size=(populationlength, botlength))

# Вывод размера популяции
print(f'Размер популяции: {population.shape}')

# Вывод случайного бота
print(f'\nСлучайный бот: \n{population[np.random.choice(populationlength)]}')
```

```python
# Запускаем цикл обучения
for epoch in range(1, epochs):
    # Уменьшаем мутацию
    if epoch %200 == 0:
        mutation*= 0.5
    # Расчет ошибок ботов
    val = np.array(tuple(map(fbybot, population)))

    # Получение лучших ботов из популяции
    newpopul, sval = getsurvpopul(val)

    # Выводим 5 лучших ботов
    if epoch %10 == 0:
        print(f'{epoch}   {[round(s,8) for s in sval[0:5]]}')

    # Генерируем случайные пары родителей и индекс для каждого элемента
    parents = np.array([
                           np.random.choice(
                               np.random.randint(0, nsurv, 2), 
                               botlength) 
                           for _ in range(nnew)]).reshape(-1)
    
    # Создаем массив индексов для индексации новой популяции
    indexes = np.vstack((
        parents,
        np.tile(np.arange(botlength), nnew)
    ))

    # Создаем новую популяцию
    curpopul = newpopul[tuple(map(tuple, indexes))]

    # Генерируем индексы для выборочной мутации
    indexesmut = np.random.choice(
        [False, True],
        size = curpopul.shape[0],
        p=[1-mutation, mutation])
    
    # Производим выборочную мутацию по сгенерированным индексам
    curpopul[indexesmut] += 2*np.random.random(curpopul[indexesmut].shape) - 1.0

    # Добавляем новую популяцию
    newpopul[nsurv:] = curpopul.reshape(-1, botlength)

    # Обновляем популяцию
    population = newpopul.copy()
```

```python
# Получение ответа
x = population[0,0] + 5*population[0,1] + 10*population[0,2] + 25*population[0,3]
y = population[0,4] + 5*population[0,5] + 10*population[0,6] + 25*population[0,7]
print('**Решение**')
print(f'x={x}')
print(f'y={y}')
print()
print(f'Значение функции f(xy)={f(x,y)}')
```
