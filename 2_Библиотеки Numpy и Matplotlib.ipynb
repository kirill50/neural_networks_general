##Подключение библиотеки и простейший пример использования

Изучая синтаксис, вы можете задаться вопросом: как знание **Python** поможет при работе с нейронной сетью? Понимает ли она команды на **Python**? И да, и нет. Дело в том, что команду собрать сеть вы отдаете самому компьютеру, и тут ваше общение идет на **Python**. А вот с самой нейросетью все гораздо интереснее: она может работать как с числами и словами, так и с текстами, картинками, даже со звуками! Но как она понимает такое большое количество разных вещей, если она не человек?

Не человек, но стремится к этому и быстро всему учится. Просто вы переводите любую информацию на язык чисел, который понятен нейросети. И тут вам повезло: есть много помощников, которые облегчают задачу обучения нейросети.

Помощник номер один – библиотека **Numpy**. 

Название состоит из двух частей – **Num** (сокращение от Numerical, означает «числовой») и **Py** (сокращение от уже знакомого вам **Python**). **Numpy** преобразует все, что вы захотите, в язык чисел. Это могут быть изображения, аудиозаписи, тексты и другие данные. Язык чисел необходим для передачи сети любой информации в формате, который она воспринимает.

Вот пример того, как любая библиотека (в нашем занятии это **Numpy**) может быть привлечена к работе на ваше благо:


````python
import numpy as np
````

Буквально, на языке **Python** под фразой `import numpy as np` имеется в виду «Вызови **Numpy** под псевдонимом **np**!». Псевдонимы применяют для краткости в командах кода. Как вы уже знаете, так можно «зашифровать» все библиотеки и команды.

**Как связана библиотека Numpy с работой с числами?**

Суть работы **Numpy** в том, что с помощью этой библиотеки любую информацию можно представить в виде чисел. Чем больше информации, тем больше требуется числовых значений.
Простой пример:


````python
my_list = [1, 2, 3, 4, 5]                 # Задание списка чисел с названием my_list
numpy_arr = np.array(my_list)             # Вызов np.array() превращает список в массив чисел
print(numpy_arr)                          # Проверка результата
print(type(numpy_arr))                    # Вызов type() сообщает тип данных значения переменной numpy_arr
````

Теперь посмотрите: если видите тип данных **numpy.ndarray** (определяется вызовом `type()`), то числа в таком типе данных можно считать элементами одного целого массива (**"array"** – массив). Его можно создать при помощи **Numpy**, и это будет один цельный массив чисел.

Чтобы не повторяться, пусть теперь он называется **new_arr**:

````python
new_arr = np.arange(1, 6)                 # Вызов arange(1, 6) задает массив значений от 1 до (не включая) 6
print(new_arr)                            # Проверка результата
````

В массиве можно сделать числа просто случайными, будто наугад. Для этого нужны две команды: 
- `random()` (означает «случайный»)
- `randint()` (означает «случайный интервал»). 

Объединяя эти команды, укажем границы для значений в массиве и количество чисел. Вот что получится:

````python
# Массив будет сохранен в переменной "arr"
arr = np.random.randint(1, 11, 5)         # От 1 до (не включая) 11 - границы массива, а число 5 - общее кол-во чисел
print(arr)                                # Проверка результата
````

***Обратите внимание:*** при использовании как `arrange(1, 6)`, так и `randint(1, 11)` второе число в скобках никогда не будет включено в диапазон. Так условились между собой язык **Python** и компьютер. Запомните это на будущее.

Зачем все это для работы с нейросетью? Все просто: нейронная сеть воспринимает информацию в виде массивов чисел **Numpy**. Каждый такой массив представляет собой организованное множество числовых данных. Инструменты работы с массивами собраны в библиотеке **Numpy**.

##Типы данных в массивах Numpy

Все элементы в массивах обязательно приводятся к одному виду. То есть один массив – один тип данных. Самих же типов данных в **Numpy** несколько. Однако, если они оказываются в одном массиве, то приводятся к общему «структурному» знаменателю:
- если есть только целые числа, то к типу **int**;
- если и целые, и дробные, то все значения становятся дробными. Здесь любое целое, например **5**, просто станет **5.0**. Дробный тип называют **float**;
- если среди данных появляются слова (текст), то данные обобщаются по универсальному кодовому стандарту – **Unicode**. Но вы встретите просто сокращение – большую букву **U**.

***Совет:*** чтобы узнать тип данных конкретного массива, используйте функцию `dtype()`

Рассмотрим примеры массивов с одинаковым именем, но разным содержимым:



````python
arr = np.array([1, 2, 3, 45, 7])          # Вот, пожалуйста - массив только из целых чисел
print(arr.dtype)                          # Здесь выясняется тип данных: int64
````

Пусть цифра **64** в типе данных **int64** вас не смущает. Разные числа (**8, 16, 32, 64** и больше) можно увидеть и после типа данных **int**, и **float**, и **U**. Такая цифра лишь означает объем памяти, которую занимает единица данных. 

````python
arr = np.array([1.23, 7.1, 3.14])         # Здесь массив содержит уже дробные числа
print(arr.dtype)                          # Тип этих данных: float64
````

````python
arr = np.array(["First", "Second", "blabla", 12345])  # А вот тут добавлены разные строки (фактически - текстовые данные)
print(arr.dtype)                                      # И тип данных будет <U6, где все значения - строки длиной не более 6 символов
````

##Размерности и форма массивов **Numpy**

Для трансформации даже самых скудных данных требуется далеко не одно число. На практике даже одно-единственное число может быть массивом **Numpy**, если оно заключено в квадратные скобки `[ ]`. 
1. Если число – это просто число, то его называют ***скаляром***. Например, **4** – скаляр, **[4]** – массив.
2. Несколько чисел в линию в квадратных скобках – это вектор. Например, **[1 2]** – вектор.
3. Если есть и линия (ряд), и столбик, будто это таблица, то это матрица. Например:
    ```
    [[-4 -5]
     [ 2 -1]]
    ```
  Это все одна матрица.
4. Чаще всего вы встретите не просто матрицы, а матрицы из матриц, одно внутри другого – для наибольшей вместимости. Их именуют тензорами (что это такое – вы подробнее узнаете в следующих занятиях). Например:

    ```
    [[[-4 -5]
      [ 2 -1]]

     [[ 4 -1]
      [-4  0]]

     [[ 0 -2]
      [-4 -3]]]
    ```
  Это уже тензор.

Кстати, создать тензор со случайными числами можно все теми же методами. Не пугайтесь большого количества квадратных скобок в начале тензора и в конце. Вспомните выдвижные ящики с карточками в регистратуре или библиотеке. Тензор похож на такой ящик: его внутренние массивы располагаются по слоям.


````python
tensor = np.random.randint(-5, 5, (3, 2, 2))    # Во вторых скобках заданы размеры тензора в высоту, ширину, глубину (оси X,Y,Z)
print(tensor)                                   # Вывод трехмерного тензора
````

Свойства для определения параметров таких больших массивов
* `.ndim` сообщает, сколько измерений в тензоре (может быть и больше 3-х!);
* `.shape` сообщает ***форму массива***: насколько каждое из измерений велико в высоту/ширину/глубину (т.е. по столбцам, строкам и слоям);
* `.size` сообщает, сколько в тензоре всего занятых мест.
Посмотрим на работу методов в действии:


````python
print('Количество осей: ',tensor.ndim)
print('Форма массива: ',tensor.shape)
print('Количество значений: ',tensor.size)
````

##Индексация в массивах

Для выделения какго-либо числа в массиве воспользуйтесь поиском по индексу. Доступ к значениям под соответствующим им номерам места в массиве называется индексацией.

Найти элемент тензора можно так же, как и элемент списка **Python**:


````python
tensor = np.random.randint(-10, 10, (3, 2, 2))  # Создание трехмерного тензора
print(tensor[2])                                # Выделение одного из его слоев (слой №2 - отдельная матрица)
                                                # На место в тензоре также укажет количество квадратных скобок
````

````python
print(tensor[2, 1])                             # Выделение вектора под индексом места 1
                                                # Этот вектор берется из матрицы с индексом 2
                                                # Это видно, если сравнить результаты кода этой и предыдущей ячеек
````

````python
print(tensor[2, 1, 0])                          # Здесь способ найти отдельное число под индексом 0
                                                # Число взято из слоя 2, строки 1 и столбца 0 тензора
````

Если нужно выбрать несколько чисел по строкам матрицы в тензоре, то вы можете использовать, говоря на профессиональном сленге, ***«срез»***. 

Рассмотрим большую матрицу случайных чисел, размером **5** столбцов на **5** строк (**5 х 5**):

````python
matrix = np.random.randint(-10, 10, (5, 5))     # Значения матрицы случайны в пределах (-10, 10), 5 строк и 5 столбцов
print(matrix)                                   # Проверка результата
````

````python
print(matrix[:3])                               # Здесь знак двоеточия дает указание взять все значения первых трёх строк 
````

Если вам нужно взять из матрицы лишь одну строку, скажем, первую, то сделайте вот так:


````python
print(matrix[0])                                # А если нужно взять лишь 1-ю строку, то это делается так:
                                                # Напоминение: отсчёт индексов всегда ведётся с нуля!
````

Следующим способом можно взять только числа первого столбца. 

Для этого захватите знаком двоеточия все строки, а потом выделите в каждой строке только первый элемент (это по-прежнему индекс «ноль»). 

Вот и получится первый столбец:


````python
print(matrix[:, 0])                # Обратите внимание на конструкцию в квадратных скобках: этот способ дает первый столбец матрицы
````

````python
print(matrix[:, 1])                # А здесь выделяется второй столбец
````

При желании найти одно конкретное число сообщите его индекс в строке и индекс в столбце. Это похоже на отметку координат точки на карте. Посмотрите, где у вас в матрице элемент второй строки и третьего столбца. Только помните про отсчет с нуля, а потому ставьте значения координат на единицу меньше!

````python
print(matrix[1, 2])                # Выделение элемента второй строки третьего столбца
````

##Операции с тензорами

Для наглядности рассмотрим два небольших массива чисел:

````python
arr_1 = np.random.randint(-10, 10, 5)
arr_2 = np.random.randint(-10, 10, 5)
print(arr_1)
print(arr_2)
````

А здесь сразу ряд примеров простой математики: операции, которые вы можете выполнить с массивами. 

Действия выполняются между парой чисел, которые стоят друг под другом в предыдущей ячейке (каждом из массивов). Это поэлементные математические действия. 

Затем, чуть ниже есть строки, где вы выполните простые действия с одним массивом. Чтобы было понятнее, рассмотрите комментарии:

````python
print(arr_1 + arr_2)                     # Поэлементное сложение двух массивов
print(arr_1 * arr_2)                     # Поэлементное умножение двух массивов
print(arr_1 / arr_2)                     # Поэлементное деление первого массива на второй
print(arr_1 * 5)                         # Умножение каждого элемента массива "arr_1" на число 5
print(arr_1 / 7)                         # Деление каждого элемента массива "arr_1" на число 7
print(arr_1 ** 3)                        # Возведение каждого элемента массива "arr_1" в степень 3
````

Полезные методы (команды для массивов), которые пригодятся для статистических расчетов в процессе освоения курса:
- `.sum()` – сумма элементов;
- `.mean()` – среднее элементов;
- `.min()` – минимальный элемент;
- `.max()` – максимальный элемент.

Создадим еще один экспериментальный массив **my_array**, на котором опробуем методы:

````python
my_array = np.random.randint(-10, 10, (5, 5))   # Создание массива
print(my_array)                                 # И вывод его на экран
````

Далее все понятно и без лишних комментариев:

````python
print("Сумма всех элементов: ", my_array.sum(axis=1))
print("Среднее всех элементов: ", my_array.mean(axis=0))
print("Максимум всех элементов: ", my_array.max())
print("Минимум всех элементов: ", my_array.min())
````

***Важный момент:*** попробуйте в скобках этих команд написать слово **axis** («ось»). Если задать **axis=1**, то команда будет применяться ко всем числам каждой строки, а если **axis=0**, то ко всем числам каждого столбца:

````python
print("Сумма всех элементов по первой оси: ", my_array.sum(axis=1))
print("Среднее всех элементов по первой оси: ", my_array.mean(axis=1))
print("Максимум всех элементов по первой оси: ", my_array.max(axis=1))
print("Минимум всех элементов по первой оси: ", my_array.min(axis=1))
````

#**Matplotlib** - библиотека для визуализации данных

##Подключение библиотеки и простейшие примеры использования

Чаще всего вы будете визуализировать данные с помощью модуля данной библиотеки под названием **pyplot**:

````python
# Подключение библиотеки визуализации данных
import matplotlib.pyplot as plt  
 
# Команда для отображения графики в ячейках среды Google Colaboratory
%matplotlib inline                
````

Благодаря **Numpy** у вас всегда есть материалы в виде массивов чисел. Сейчас они пригодятся на практике.

Создадим ряд чисел одно за другим, подобно линии, с помощью команды `np.linspace()`

Числа равномерно распределяются в интервале от **0** до **15** (это будут первые два значения в скобках). А всего чисел в ряде будет **101** (третье значение в скобках). Рассмотрим подробнее:


````python
import numpy as np                        
x = np.linspace(0, 15, 101)

# Вот какой массив х получился в итоге:
print(x)
````

Немного школьной алгебры от **Numpy**:

````python
y = np.sin(x) 

# Содержимое массива у выглядит так:
print(y)              
````

Теперь это точки для построения графика синуса (значения по оси x). Для вычисления значений синуса (значений по оси y) вызовем `np.sin(x)`, где **x** – массив значений по горизонтали. При этом получится массив значений синуса, каждое из которых будет вычислено от аргумента на том же месте в массиве **x**.


**Как бы вы построили график, если у вас есть точки и для x, и для y?**

Наверное, вы бы начертили две пересекающиеся линии, которые перпендикулярны. Потом отметили место их пересечения как начало отсчета или «ноль». Потом решили, что вертикальная линия – это ваша ось **y**, горизонтальная – ось **x**, и начали бы поочередно наносить точки из массивов **x** и **y**, которые вы уже получили. Так бы получился график `sin(x)`, верно? 

Теперь это можно сделать за пару действий:


````python
plt.plot(x, y)                                # Отрисовка графика командой plot(), в которую переданы массивы точек для x и y
plt.show()                                    # С помощью .show() выводится график
````

Если вам необходимо добавить на график подписи, то это в разы легче, чем написать их в тетради.

В этом вам помогут дополнительные команды:

````python
plt.plot(x, y)                                # Отрисовка того же графика
plt.title("y=sin(x)")                         # Команда .title() - и график подписан указанным в скобках именем
plt.xlabel('Переменная X')                    # Команда .xlabel() добавит надпись только для оси x
plt.ylabel('Переменная Y')                    # А команда .ylabel() добавит надпись для оси y
plt.show()                                    # Фиксация результата и вывод - командой .show()
````

Также вы можете сделать график удобного для вас размера. Пространство, где он рисуется, называется полотном графика. Для указания размера полотна используйте команду `.figure()`. Не забудьте указать в скобках слово-параметр **figsize=** и в кортеже сами размеры для вашего графика:


````python
plt.figure(figsize=(10, 5))                   # И вот - размер полотна в дюймах указан
                                              # Первое число 10 - по оси x, второе число 5 - по оси y 
````

Несложно нарисовать и больше одного графика на одном и том же полотне. У вас часто будет возникать такая необходимость. Для этого продолжайте использовать команду `.plot()`. Не забудьте указать слегка видоизмененные данные, чтобы графики не друг с другом. Здесь используем обычную функцию из значений `cos(x)/2`, которая имеется в **Numpy**:


````python
plt.plot(x, y, label='sin(x)')                # Параметр label поможет подписать кривую графика (см. далее)
plt.plot(x, np.cos(x)/2, label='cos(x)/2')    # А так добавляется еще один график на полотно
````

Соберем все способы вместе, чтобы посмотреть на результат.

Для большей информативности добавим команды подписей графика `.title()`, `.xlabel()`, `.ylabel()`:

````python
plt.figure(figsize=(10, 5))                   # Здесь задаётся размер полотна в дюймах
                                              # по оси x и y соответственно

plt.plot(x, y, label='sin(x)')                # Эта команда строит график, а параметр label задает подпись кривой
plt.plot(x, np.cos(x)/2, label='cos(x)/2')    # Эта команда строит еще один график cos(x) на полотне

plt.title("y = sin(x)")                       # Заголовок графика
plt.xlabel('Переменная X')                    # Надпись для оси x
plt.ylabel('Переменная Y')                    # Надпись для оси y

plt.legend()                                  # А вот и новая команда для "легенды" - legend()
                                              # Она позволяет отобразить подписи кривых, указанные в .plot(label=)

plt.show()                                    # Команда для воплощения графика в жизнь :-)
````

Если для удобства восприятия графика вам нужна сетка на полотне, используйте метод `.grid()`:

````python
plt.figure(figsize=(10, 5))                   # Задание размера полотна в дюймах

plt.plot(x, y, label='sin(x)')                # Параметр label задает подпись кривой
plt.plot(x, np.cos(x)/2, label='cos(x)/2')    # Отрисовка еще одного графика на полотне

plt.title("y = sin(x)")                       # Заголовок графика
plt.xlabel('Переменная X')                    # Надпись для оси x
plt.ylabel('Переменная Y')                    # Надпись для оси y

plt.legend()                                  # Отрисовка подписей кривых ("легенды")

plt.grid()                                    # И только здесь новое - команда grid(), чтобы отрисовать сетку на фоне

plt.show()                                    # Вывод графика
````

##Слова-параметры для улучшения стиля графика##

Существует очень много различных слов-параметров, помогающих улучшить вид вашего графика. Их можно дополнительно указать в скобках команд, с которыми вы уже познакомились. Дополнительные параметры могут значительно изменить вид вашего графика.
Посмотрим на вариант графика, построенного с дополнительными параметрами:


````python
plt.figure(figsize=(14, 7))                   # Задание размера полотна в дюймах

plt.plot(x,y , label='sin(x)', color='#4b0082', linewidth=4,
         marker='h', markerfacecolor='lightgreen', markeredgewidth=2,
         markersize=10, markevery=5)

plt.plot(x, np.cos(x)/2, label='cos(x)/2', linewidth=3, color='lightcoral',
         marker='D', markeredgecolor='black', markevery=5)

plt.title('y = sin(x)')                       # Заголовок графика
plt.xlabel('Переменная X')                    # Надпись для оси x
plt.ylabel('Переменная Y')                    # Надпись для оси y

plt.legend()                                  # Отрисовка названий графиков ("легенды")

plt.grid()                                    # Отрисовка сетки

plt.show()                                    # Вывод графика
````

Как вам график?

Наверное, он кажется вам как минимум более экзотичным. Но он же будет еще и просто удобнее в работе. Рассмотрим новые слова, которые использовались выше в скобках команд. Отметьте для себя, какие вы уже видели, а какие нет:
- **marker** – маркер;
- **color** – цвет линии;
- **mec** – цвет маркера;
- **linestyle** – стиль линии;
- **label** – подпись кривой графика;
- **linewidth** – ширина линии;
- **markersize** – размер маркера.

Другие команды вы найдете в полном описании документации **Matplotlib**. Описанные там элементы стиля дают возможность существенно улучшить наглядность графика.
