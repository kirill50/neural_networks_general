###Пример использования


```python
# Инициализация клиента
client = texttospeech.TextToSpeechClient()

# Получаем список всех русских голосов
voices = client.list_voices(language_code='ru-RU')
```

Каждый голос имеет следующие параметры:

    language_codes - код языка (для русского ru-RU)
    name - название голоса
    ssml_gender - пол
    natural_samlple_rate_hertz - частота дискретизации голоса в Гц

```python
# Выведем список голосов
print(voices)
```

Попробуем озвучить стихотворение А. Блока.

```python
# Вводим текст

text = """Ночь, улица, фонарь, аптека,
          Бессмысленный и тусклый свет.
          Живи ещё хоть четверть века —
          Всё будет так. Исхода нет.

          Умрёшь — начнёшь опять сначала
          И повторится всё, как встарь:
          Ночь, ледяная рябь канала,
          Аптека, улица, фонарь."""

# Отправляем текст в функцию SynthesisInput
text_input = texttospeech_v1.types.SynthesisInput(text=text) 
```

Выбираем любой голос из списка выше.

```python
# Выбор голоса

voice = texttospeech_v1.types.VoiceSelectionParams(
    language_code='ru-Ru',
    name = 'ru-RU-Standard-B')
```

Зададим параметры аудиозаписи, которую получим на выходе.

```python
# Конфигурация аудио

audio_config = texttospeech_v1.types.AudioConfig(
    audio_encoding=texttospeech_v1.AudioEncoding.MP3, # Формат мр3
    speaking_rate = 0.8)                              # Скорость произношения
```

Отправляем текст, голос и конфигурацию аудио в функцию synthesize_speech() и получаем синтезированную речь.

```python
# Синтезируем речь

response = client.synthesize_speech(input=text_input,   # Текст синтеза
                                    voice=voice,        # Параметры голоса
                             audio_config=audio_config) # Конфигурация аудио
```

```python
# Ответ сохраняем mp3 формате

with open('output.mp3', 'wb') as out:
    out.write(response.audio_content)
    print('Файл сохранен под именем output.mp3')
```

```python
# Воспроизведем полученную аудиозапись
Audio('output.mp3')
```

Данная система синтеза речи может использовать разные модели. В предыдущем примере синтеза речи мы использовали стандартную систему (это прописано в имени голоса 'ru-RU-Standard-B'). Эта система наиболее простая, поэтому имеет не самое высшее качество (речь не идеально похожа на человеческую). Рассмотрим модель, которая синтезирует речь на основе wavenet (wavenet синтезирует премиальные голоса).

### Отличие Wavenet от базовых голосов

```python
# Зададим параметры голоса
voice = texttospeech_v1.types.VoiceSelectionParams(
    language_code='ru-Ru',
    name = 'ru-RU-Wavenet-E') # Укажем модель Wavenet

# Конфигурация аудио
audio_config = texttospeech_v1.types.AudioConfig(
    audio_encoding=texttospeech_v1.AudioEncoding.MP3, # Формат мр3
    speaking_rate = 0.9)                              # Скорость произношения

# Синтезируем речь
response = client.synthesize_speech(input=text_input,   # Текст синтеза
                                    voice=voice,        # Параметры голоса
                             audio_config=audio_config) # Конфигурация аудио

with open('output.mp3', 'wb') as out:                   # Сохраняем аудио в файл
    out.write(response.audio_content)
```

```python
# Воспроизведем полученную аудиозапись

Audio('output.mp3')
```

Для сравнения возьмём те же самые параметры голоса, но со стандартной моделью.

```python
# Standard
voice = texttospeech_v1.types.VoiceSelectionParams(
    language_code='ru-Ru',
    name = "ru-RU-Standard-E") # Укажем стандартную модель

# Конфигурация аудио
audio_config = texttospeech_v1.types.AudioConfig(
    audio_encoding=texttospeech_v1.AudioEncoding.MP3, # Формат мр3
    speaking_rate = 0.9)                              # Скорость произношения

# Синтезируем речь
response = client.synthesize_speech(input=text_input,   # Текст синтеза
                                    voice=voice,        # Параметры голоса
                             audio_config=audio_config) # Конфигурация аудио
with open('output.mp3', 'wb') as out:                   # Сохраняем аудио в файл
    out.write(response.audio_content)
```

```python
# Воспроизведем полученную аудиозапись
Audio('output.mp3')
```

###Конфигурация аудио

Данная библиотека позволяет настраивать такие параметры аудио, как:

*  Скорость произношения
*  высота голоса
*  Частота дискретизации
*  Громкость

```python
# Будем использовать женский Wavenet голос

voice = texttospeech_v1.types.VoiceSelectionParams( 
    language_code='ru-Ru',
    name = 'ru-RU-Wavenet-E')
```

```python
# Конфигурация аудио

audio_config = texttospeech_v1.types.AudioConfig(
    audio_encoding=texttospeech.AudioEncoding.MP3, # Формат мр3
    speaking_rate = 0.9,                           # Скорость произношения
    pitch = 0,                                     # Высота голоса
    sample_rate_hertz = 24000,                     # Частота дискретизации
    volume_gain_db = 0)                            # Усиление громкости, дБ
```

Разная высота голоса:

```python
audio_config = texttospeech_v1.types.AudioConfig(
    audio_encoding=texttospeech.AudioEncoding.MP3, # Формат мр3
    speaking_rate = 0.9,                           # Скорость произношения
    pitch = -5)                                    # Высота голоса

response = client.synthesize_speech(input=text_input,   # Текст синтеза
                                    voice=voice,        # Параметры голоса
                             audio_config=audio_config) # Конфигурация аудио
with open('output.mp3', 'wb') as out:
    out.write(response.audio_content)
```

```python
# Воспроизведем полученную аудиозапись

Audio('output.mp3')
```
