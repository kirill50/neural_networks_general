# Pandas

## Знакомство с библиотекой

Как вы уже знаете, нейронные сети работают с числами - получают их на вход и выдают на выходе. Для преобразования любой информации в язык чисел мы пользуемся библиотекой NumPy.

Но перед тем как работать с данными, их нужно проанализировать, обработать. И как раз в этом нам поможет библиотека <a href="http://pandas.pydata.org/">**Pandas**</a>. Она претендует на звание самого мощного и гибкого средства для анализа данных с открытым исходным кодом.

И как обычно, начнем с подключения библиотеки:

```
# Загрузка модуля pandas
import pandas as pd
```

С помощью библиотеки **Pandas** вы обработаете табличные данные, что схоже с работой в Excel. В вашем распоряжении две структуры данных:

- **DataFrame** - двухмерный массив, имеет табличную структуру, легко изменяется по размерам, может содержать в себе данные разных типов
- **Series** - одномерный массив с именованными индексами (чаще всего, данные одного типа).

DataFrame - это таблица. Она состоит из столбцов. Каждый такой столбец - это структура Series.

Оба типа можно создавать вручную с помощью функций из самой библиотеки:
- `pandas.Series(data=None, index=None, dtype=None)`
- `pandas.DataFrame(data=None, index=None, columns=None, dtype=None)`

Аргументы:
- **data** - данные, которые необходимо записать в структуру
- **index** - индексы строк
- **columns** - названия столбцов
- **dtype** - тип данных

## Создание датасета

А теперь попробуйте самостоятельно воспроизвести датасет:

<img src="https://i.imgur.com/FUCGiKP.png">

Первое, что вам нужно сделать, это создать списки, содержащие названия столбцов, индексы и сами данные:

```
# Создание списка с названиями столбцов
columns = ['country', 'province', 'region_1', 'region_2']
```

Вовсе необязательно проставлять индексы в списке, поскольку они автоматически нумеруются, начиная с нуля: 0, 1, 2, 3... и так далее. Но в таблице, которую вы воссоздаете, индексы указаны не по порядку. Поэтому вам нужно самостоятельно создать список с нужными индексами.

Помните, что индекс нужен для того, чтобы идентифицировать каждую строчку. Поэтому они не должны повторяться:

```
# Создание списка с индексами строк
index = [0, 1, 10, 100]
```

Отлично! Следующий шаг – создание списка с данными. Список будет двумерным, таким образом каждая строка таблицы идет отдельным списком:

```
# Создание списка с данными ячеек
data = [['Italy', 'Sicily & Sardinia', 'Etna', 'NaN'], 
        ['Portugal', 'Douro', 'NaN', 'NaN'],
        ['US', 'California', 'Napa Valley', 'Napa'],
        ['US', 'New York', 'Finger Lakes', 'Finger Lakes']]
```

Пока что вы создали три переменные, которые не имеют никакой связи между собой. 

Чтобы связать эти данные и образовать датафрейм, используйте метод `.DataFrame`.

Этот метод принимает следующие параметры:
- список с данными;
- columns - список с названиями столбцов;
- index - список с индексами.

Теперь создайте датафрейм и выведите его на экран:


```
# Создание датафрейма (таблицы). В качестве параметров передаются названия столбцов, индексы и сами данные
df = pd.DataFrame(data, columns=columns, index=index)

# Отображение таблицы
df
```

Как вы знаете, индексы можно и не указывать. Попробуйте убрать этот параметр:


```
# Создание датафрейма. В качестве параметров используем только названия столбцов и данные
df = pd.DataFrame(data, columns=columns)

# Отображение таблицы
df
```

Правильно, индексы были созданы автоматически! Если вы все-таки решите вручную создать индексы, будьте внимательны. Если количество элементов в списке с индексами не будет соответствовать количеству строк таблицы, компилятор выдаст ошибку.

Вы можете заметить, что функция `print()` не использована для вывода таблицы. Чтобы понять, почему, попробуйте отобразить датафрейм с помощью этой функции:


```
print(df)
```

Выглядит не очень красиво. Запомните эту особенность и впредь не используйте `print()` для отображения датафреймов. Вместо этого либо просто пишите название датафрейма в последней строке, либо воспользуйтесь функцией `display()`. Особенно удобно ее применять в случаях, когда нужно вывести несколько датафреймов:

```
display(df, df)
```

Или так:

```
print('Первый датафрейм:')
display(df)
print('Второй датафрейм:')
display(df)
```

Вот и все! Таким образом вы самостоятельно создали датасет. Но такой подход вы редко будете использовать. Куда чаще вам нужно будет создавать датафреймы из уже готовых файлов.

Чтобы это исправить, при создании датафрейма укажите параметр `index_col=0`. Теперь информация об индексах будет взята из нулевого столбца. Проверьте результат:

Вы обнаружили пропуски, но что с ними делать? Ячейки без информации можно удалять и заполнять самостоятельно. Рассмотрите подробнее каждый вариант.

**1. Удалять**

Команда `dropna(axis=0, how='all')` ничего не удалит, так как полностью пустых строк нет.

По тому же принципу можно вывести каждую пятую строку. То есть вывести все строки с шагом в пять:

Метод `.loc` также позволяет отбирать данные по строкам. На самом деле его можно применять более функционально, но об этом позже.

В случае с `.loc` принцип работы такой же. Указываете индексы и захватываете определенные диапазоны. Разница лишь в том, что последний индекс включен в выборку:

```
# Отбор данных строк с 10 до 13 (включительно)
data.loc[10:13]
```

Точно так же можете вывести диапазон строк с определенным шагом:

И на выходе снова имеете два идентичных способа записать данные строк в переменную:


```
array3 = data[10:12]
print(array3)

print(end='\n\n\n')

array4 = data.loc[10:12]
print(array4)
```

#### Отбор по столбцам и строкам

Вы узнали, что метод `.loc` можно использовать более функционально, чем отбирать только строки. Речь о том, что с помощью этого метода можно одновременно отбирать и строки, и столбцы. Делают это следующим образом:


```
# Отбор данных строк с 10 по 12, столбцов 'price' и 'points'
data.loc[10:12, ['price', 'points']]
```



---


**Важно**: метод `.loc` не позволяет обращаться только к столбцам. То есть с его помощью вы обращаетесь либо к строкам, либо к строкам и столбцам. Но не к одним только столбцам!


---



Есть немного измененная версия изученного метода – это `.iloc`. Он решает те же задачи, но обращение к столбцам идет не по названиям, а по индексу. Нумерация такая же, как в NumPy-массивах: 0, 1, 2, 3...


```
# Отбор данных строк с 10 по 12, столбцов 4 и 3
data.iloc[10:12, [4, 3]]
```

Метод `.iloc`, как и .loc, позволяет обращаться только к строкам, но не позволяет обращаться только к столбцам.

Чтобы запомнить, за что отвечает каждый метод: приставка `i` у второго метода происходит от слова `index`. Значит, если приставка есть, то метод обращается к столбцам по индексу. А если приставки нет, то по названию.


#### Отбор с условиями

Часто появляется необходимость отделить данные по каким-либо показателям из самого датафрейма. Например, отобрать данные по всем винам дороже 1 000 $. Рассмотрите два способа, которые помогут решить эту задачу.

Первый способ: **маска**

Суть в том, что вы создаете переменную-маску, которая содержит в себе информацию о том, какие вина проходят по условию, а какие – нет. Чтобы было понятнее, рассмотрите пример. 

Создайте переменную-маску следующим образом:


```
# Создание маски
mask = data['price'] > 1000
```

То есть вы записываете в переменную `mask` результат логического выражения 

        data['price'] > 1000. 
        
Другими словами, вы берете столбец `price` из датафрейма `data` и проверяете, какие значения больше 1 000.

Теперь `mask` содержит информацию в следующем виде:


```
# Отображение маски
mask.head()
```

Каждая строка приняла значение **False** или **True**:
- True – если стоимость вина в строке дороже 1 000 долларов;
- False – если стоимость вина в строке дешевле или равна 1 000 долларов.

Используйте `.head()`, чтобы вывести только первые пять строк. Это позволяет не нагромождать рабочее пространство визуальным шумом.


Осталось создать переменную для новой таблицы и записать в нее датафрейм с применением маски:

```
# Отбор данных в соответствии с маской в новую таблицу temp и вывод таблицы
temp = data[mask]
temp
```

Готово! Теперь вы знаете, что всего 10 вин стоят дороже 1 000 $, и большая часть из них произведена во Франции. Теперь вы можете анализировать данные по этому датафрейму, изучая именно дорогие вина.

Второй способ: **логическое условие**

Благодаря второму способу вы можете не объявлять лишние переменные и отбирать данные по более сложным условиям. Но для начала изучите азы. Отберите данные по цене, как и в прошлом способе:

```
# Цена больше 300
data[data.price > 300].head()
```

Вы сделали это всего в одну строку! Принцип действия такой же: в квадратных скобках указывается маска **цена > 300**. Если условие выполняется - **True**, если нет - **False**.

Но это еще не все. Теперь отберите все вина дороже 200 $, произведенных либо в США, либо во Франции:

```
# Отбор по составному условию
data[(data.price > 200) & ((data.country == 'US') | (data.country == 'France'))].head(10)
```

Разберем подробнее.

- **&** - **and** - логическое "и". Возвращает истину, если оба условия верны.
- **|** - **or** - логическое "или". Возвращает истину, если хотя бы одно из условий верно. 

---

- `data.price > 200`- обращаемся к столбцу **price**. Если значение больше 200, будет возвращено **True**. Если меньше или равно 200, будет возвращено **False**.
- `data.country == 'US'` - обращаемся к столбцу country. Если значение равно **US**, будет возвращено **True**, в противном случае будет возвращено **Flase**.

---

Теперь разберем само логическое выражение:
- `(data.price > 200) & ((data.country == 'US') | (data.country == 'France'))`
- **цена больше 200** `и` **страна US** `или` **страна France**

Таким образом, на выходе получим все вина дороже 200$ со страной производства или US, или France.

На данный момент вам неизвестны инструменты, с помощью которых можно решить такую задачу. Рассмотрите их!


Метод `.groupby()` позволяет провести сортировку по столбцам. Укажем в качестве параметра столбец **country** и применим метод `.count()`, чтобы вычислить количество заполненных элементов по столбцам. Полученный результат запишите в переменную **data_** и выведите первые 20 строк:

Задача выполнена! Теперь вы можете оценить базу данных с новой стороны. Например, из этой таблицы следует, что колонка region_2 заполнена только для US.
Не останавливайтесь на этом.

### Изменение значений в таблице

Сейчас вы научитесь менять данные в датафрейме и рассмотрите два способа – с помощью метода `.iloc` и метода `.loc`. 

Но сначала создайте бэкап датафрейма (копию оригинала), чтобы можно было отменить изменения в любой момент. Для этого используйте метод `.copy()`. Запишите бэкап в переменную `data_backup`:


```
# Создание копии таблицы
data_backup = data.copy()
data.head()
```

Возможно, у вас появился вопрос: почему нельзя использовать запись `data_backup = data`. Зачем нужен метод `.copy()`?

Дело в том, что если присвоить переменной `data_backup` значение переменной `data`, то при изменении `data` изменится и `data_backup`. Тут на помощь и приходит метод `.copy()`. Обратите на это особое внимание.

Делать бэкап датафрейма – хорошая привычка, которая не раз сэкономит вам уйму времени. Теперь вы можете перейти к способам изменения данных в датафрейме.


Первый способ: метод `.iloc`

Вы уже разбирали этот метод в разделе про срезы данных. Принцип тот же, только в этот раз вы приравниваете к выбранным ячейкам значения, на которые хотите поменять оригинальные записи. Выглядит это так:


```
data.iloc[0,1] = 'good day'   # Запись нового значения в 0 строку и 1 стоблец
data.iloc[2,2] = '129'      # Запись нового значения во 2 строку и 2 стоблец
data.iloc[3:5,2:5] = 'new'  # Запись нового значения с 3 по 5 строку и со 2 по 5 стоблец
data.head(8)
```

Второй способ: `.loc`

С помощью этого метода можно задавать условия, по которым будет определяться, для каких ячеек нужно изменить значение. 

Чтобы работать с исходным файлом, вернитесь к оригинальной версии датафрейма, используя все тот же метод `.copy()`:

```
# Восстановление данных из копии
data = data_backup.copy()
```

Рассмотрим работу с методом `.loc` на следующем примере: `data.loc[data.country == 'US', 'region_2'] = 'Syberia'`

- `data.loc` - используем метод на нашем датафрейме
- `data.country == 'US'` - задаем условие. В данном примере столбец **country** должен хранить значение **US**. Если это условие выполняется, данные будут перезаписаны
- `'region_2'` - дальше определяем в каком столбце будем менять данные. В примере это столбец **region_2**
- `= 'Syberia'` - в конце указываем на какое значение мы хотим поменять исходное.

Запись можно расшифровать так: для всех вин произведенных в **US** нужно записать в столбец **region_2** значение **Syberia**. Проверим код:

```
data.loc[data.country == 'US', 'region_2'] = 'Syberia'  # Запись новых значений в столбец "region_2" строк "US"
data.head()
```

Все сработало именно так, как мы задумали. Решим еще 2 задачи:

- приравнять значение **points** к 200 для всех вин дороже 100$

- приравнять значение **price** к 1000 для всех вин дороже 100$

```
data.loc[data.price > 100, 'points'] = 200              # Запись новых значений в столбец "points" для вин дороже 100
data.loc[data.price > 100, 'price'] = 1000              # Запись новых значений в столбец "price" для вин дороже 100
data.head()
```

Вот и все! Большой блок по анализу данных подошел к концу. Можете сделать перерыв, а чуть позже – вернуться и закончить урок по библиотеке **Pandas**.

## Объединение таблиц по строкам (конкатенация)

В первом уроке курса вы уже читали о конкатенации – сложении строк.

Объявите две переменные и присвойте им текстовые значения:


```
a1 = 'Нейро'
a2 = 'сети'
```

Теперь сложим 2 эти строки (это и есть конкатенация):

```
a1 + a2
```

Таким же образом происходит конкатенация списков:

```
a2 = [0, 1, 2, 3]                          # Создание списка а2
b2 = [4, 5, 6, 7, 234, 23]                 # Создание списка b2
b2 + a2 + b2 + a2 + a2                     # Конкатенация списков
```

А можно ли так же производить конкатенацию таблиц? Чтобы проверить это, создайте две маленькие таблицы. Для этого скопируйте строки из датафрейма **data**:

```
data_1 = data[0:5].copy()                # Копирование первых 5 строк таблицы data в переменную data_1
data_2 = data[5:10].copy()               # Копирование 5-10 строк таблицы data в переменную data_2
```

По индексам можно заметить, что теперь таблицы следуют в другом порядке.

Замена значения в таблице `data_1` не привела к изменениям в таблице `df_concat`. То есть метод `.concat()` создает таблицу, независимую от оригиналов.

Чтобы превратить столбец **data** в индекс типа **DateTime**, используйте  метод `.to_datetime()`:

```
# Превращение столбца date в индекс
dataTraining.index = pd.to_datetime(dataTraining.date)
dataTraining.head()
```

Информация в таблице после этого дублируется. Чтобы это исправить, удалим столбец **date** методом `.drop()`:

- `'date'` - название колонки, которую хотим удалить
- `axis=1` - удаляем столбец, а не строку
- `inplace=True` - если **True**, то столбец удалится в исходном датафрейме. Если **False**, то для сохранения изменений нужно записать данные в переменную.

Чтобы было понятнее, разберем последний параметр на примере.

Если мы хотим создать новый датафрейм - такой же, как dataTraining, но без столбца **date**, то мы пишем следующие команды: `newDataTraining = dataTraining.drop('date', axis=1)`. Параметр `inplace` будет равен **False** по умолчанию. В итоге у нас будет два датафрейма: **dataTraining** со столбцом **data** и **newDataTraining** без столбца **data**.

Но если наша задача состоит в том, чтобы обновить уже имеющийся датафрейм и не создавать новый, мы присваиваем параметру `inplace` значение **True**. В таком случае столбец **data** будет удален из исходного датафрейма **dataTraining**.

```
# Удаление столбца date из таблицы
dataTraining.drop('date', axis=1, inplace=True)
dataTraining.head()
```

Цель достигнута! Теперь столбец **date** является индексом таблицы.

**Матрица корреляций** `.corr()`

Этот метод выведет матрицу с коэффициентами корреляции Пирсона между данными всех столбцов. Она показывает как сильно элементы таблицы коррелируют друг с другом.
- **1** - параметр полностью коррелирует (н-р, температура растет, освещенность растет)
- **-1** - параметр коррелирует в противоположную сторону (н-р, влажность растет, температура падает)
- **0** - параметр не коррелирует (между изменениями температуры и освещенности нет никакой связи)

```
# Вывод матрицы корреляции для таблицы dataTraining
dataTraining.corr()
```

**Кумулятивные функции** `.cumsum()`, `.cummin()` и `.cumprod()`

По-другому эти функции называют накапливающими. Разберем принцип действия этих функций на примерах. Чтобы было удобнее ориентироваться, выведем исходную таблицу.

```
# Вывод первых 5 строк таблицы dataTraining
dataTraining.head()
```

Начнем с кумулятивной суммы.

- Первая строчка соответствует строке оригинального датафрейма.
- Вторая строка является результатом сложения значений первой и второй строки оригинального датафрейма.
- Соответственно, третья строка является результатом сложения значений первой, второй и третьей строк оригинала и так далее.

```
# Вывод кумулятивной суммы
dataTraining.cumsum().head()
```

По такому же принципу можно вычислить кумулятивный минимум. 

- Первая строка соответствует оригиналу.
- Во второй строке сравниваются значения для каждого столбца из первой и второй строки и в результат записываются минимальные значения.
- В третьей строке сравниваются значения из всех предыдущих строк и записывается минимальное и так далее.

```
# Вывод кумулятивного минимума
dataTraining.cummin().head()
```

Кумулятивное произведение.

- Первая строка соответствует исходному датафрейму.
- Вторая строка хранит результаты умножения значений первой строки на значения второй строки.
- И так далее.

```
# Вывод кумулятивного произведения
dataTraining.cumprod().head()
```

**Метод `.apply()`**

Этот метод позволяет применить к датафрейму самостоятельно созданные функции.

Чтобы методу `.apply()` было с чем работать, создайте собственную функцию для нормирования значений

Нейросети значительно лучше работают с нормально распределенными данными. Это значит, что у набора числовых данных среднее значение должно равняться нулю, а среднее квадратичное отклонение – единице.

Значит, вам нужно написать такую функцию, которая преобразует данные в соответствии с этим правилом:


```
# Создание функции для нормирования значений
def scaling(df):
    return (df - df.mean()) / df.std()
```

Вы создали функцию для нормирования значений. Теперь с помощью метода `.apply()` примените ее к датафрейму:

```
# Применение созданной функции к таблице dataTraining (по умолчанию по столбцам)
dataTraining.apply(scaling).head()
```

**lambda-функции и метод `.agg()`**

Допустим, вы хотите по порядку вывести сумму значений всех элементов в столбце, среднее значение, минимальное и максимальное. Вы уже знаете, что это можно сделать, выписав по отдельности команды `dataTraining.sum()`, `dataTraining.mean()` и так далее. Но можно применить более удобный и компактный способ.

Метод `.agg()` позволяет объединить вызов разных команд и даже собственных функций следующим образом:

```
# Вычисление требуемых значений по всем столбцам dataTraining
dataTraining.agg(['sum', 'mean', 'min', 'max', lambda x: x.mean()])
```

`lambda` - это одноразовая функция. Их также называют **анонимными функциями**. В данном случае с помощью нее вычисляются средние значения, поэтому результат соответствует результату команды **mean**.

**Ресэмплинг** `.resample()`

Этот метод позволяет легко произвести подсчеты по дням, месяцам, годам. Например, если перед вами стоит задача вывести среднее значение температуры по дням:

```
dataTraining.Temperature.resample('D').mean()
```

В данном случае **D** означает Day - день. Разберем некоторые другие возможные варианты:

- D - day - день
- W - week - неделя
- M - month - месяц
- H - hour - час
- T - minute - минута
- S - second - секунда

Для закрепления материала выведите максимальные значения температуры по неделям:

```
dataTraining.Temperature.resample('W').max()
```

Или данные о максимальной температуре за периоды по 3 дня:

```
dataTraining.Temperature.resample('3D').max()
```
